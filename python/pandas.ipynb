**FILTERING,SORTING AND GROUPING**

```python
#FILTERING

import pandas as pd

def filter_mileage(df):
    df['MILEAGE'] = pd.to_numeric(df['MILEAGE'])
    filtered_df = df[df['MILEAGE'] > 30]
    return filtered_df


filtered_data = filter_mileage(df)
print(filtered_data)

```

```python
#SORTING

sorted_df = df.sort_values(by='MILEAGE')
sorted_df_desc = df.sort_values(by='MILEAGE', ascending=False)

sorted_df
```

```python
#GROUPING

df['MILEAGE'] = pd.to_numeric(df['MILEAGE']).fillna(0).astype(float)
grouped = df['MILEAGE'].groupby(df['COMPANY'])  

print(grouped.mean()) 


```

**MERGING/JOINING**

```python
#merging

import pandas as pd

data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],}

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']}

dd = pd.DataFrame(data1)

dd1 = pd.DataFrame(data2)

res = pd.merge(dd, dd1, on='key')

res
```

```python
#joining

import pandas as pd


data1 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32]}

data2 = {'Address':['Allahabad', 'Kannuaj', 'Allahabad', 'Kannuaj'],
        'Qualification':['MCA', 'Phd', 'Bcom', 'B.hons']}

df = pd.DataFrame(data1,index=['K0', 'K1', 'K2', 'K3'])

df1 = pd.DataFrame(data2, index=['K0', 'K2', 'K3', 'K4'])

res = df.join(df1)

res

```

```python
res1 = df.join(df1, how='outer')

res1
```

**HANDLING MISSING VALUES**

```python
#finding null value

print(df.notnull())
```

```python
#Using dropna()
import pandas as pd

def drop_missing_rows(df):
    dfc = df.copy()
    dfc.dropna(inplace=True)
    return dfc


print(drop_missing_rows(df))
   
```

```python
#using fill
def fill_missing_rows(dc):
    dcc = dc.copy()
    dcc.fillna(0, inplace=True)  
    return dcc

print(fill_missing_rows(dc))
```

```python
#Using ffill and bfill

def ffill_missing_rows(dc):
    dcc = dc.copy()
    dcc.ffill(inplace=True)
    return dcc

print(ffill_missing_rows(dc))
```

```python
#using bfill
def bfill_missing_rows(dc):
    dcc = dc.copy()
    dcc.bfill(inplace=True)
    return dcc

print(bfill_missing_rows(dc))
```

**AGGREGATION**

```python

df['CC'] = pd.to_numeric(df['CC'], errors='coerce').fillna(0).astype(float)
df['MILEAGE'] = pd.to_numeric(df['MILEAGE'], errors='coerce').fillna(0).astype(float)
df['PRICE'] = pd.to_numeric(df['PRICE'], errors='coerce').fillna(0).astype(float)


aggregation = df.groupby('COMPANY').agg({
    'CC': ['mean', 'min', 'max'],
    'MILEAGE': ['mean', 'min', 'max'],
    'PRICE': ['mean', 'min', 'max'],
})

print(aggregation)
```
