VARIABLES, LOOPS & FUNCTION

```python
numbers = [1, 2, 3, 4, 5, 6]

def count_evens(nums):
    even = []
    count = 0
    for n in nums:
        if n % 2 == 0:
            count += 1
            even.append(n)
    return count, even

count, even_numbers = count_evens(numbers)
print("Count of even numbers =", count)
print("Even numbers =", even_numbers)
```

DATA STRUCTURES


```python
fruits = ['kiwi', 'tomato', 'apple', 'apple', 'banana', 'apple', 'cherry', 'banana']
fcount = {}
for fruit in fruits:
    if fruit in fcount:
        fcount[fruit] += 1
    else:
        fcount[fruit] = 1
print("Fruit counts:", fcount)
```

```python
unique_fruits = set(fruits)
print("Unique fruits:", unique_fruits)
```

ERROR HANDLING

```python
def safe_division():
    try:
        filename = input("Enter the filename to read numbers from: ")
        with open(filename, 'r') as file:
            numbers = file.readlines()

        num1 = int(numbers[0].strip())
        num2 = int(numbers[1].strip())

        result = num1 / num2
        print(f"Result of dividing {num1} by {num2} is {result}")

    except FileNotFoundError:
        print(f"Error: The file '{filename}' does not exist.")
    except IndexError:
        print("Error: The file does not contain enough lines (need at least two numbers).")
    except ValueError:
        print("Error: The file should contain valid integers on the first two lines.")
    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


safe_division()
```

READING & WRITING CSV/JSON

```python
# CSV write

import csv

with open('data.csv', 'w', newline='') as f:
    write = csv.writer(f)
    write.writerows(['name', 'age'])
```

```python
#csv read

with open('data.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row)
```

```python
# JSON write
import json

data = {"name": "Alice", "age": 30}
with open('data.json', 'w') as f:
    json.dump(data, f)
```

```python
#json write

with open('data.json', 'r') as f:
    d = json.load(f)
    print(d)
```
