```python
import boto3
import pandas as pd
import io

session = boto3.Session(region_name='us-east-1')
s3_client = session.client("s3")

bucket_name = "qub0013"
obj = s3_client.get_object(Bucket=bucket_name, Key='demo/data2.xlsx')
body_bytes = obj['Body'].read()                
xls = pd.ExcelFile(io.BytesIO(body_bytes))   

def exl_to_csv(xls):
    for sheet_name in xls.sheet_names:
        df = pd.read_excel(xls, sheet_name=sheet_name)
        df.to_csv(f'{sheet_name}.csv', index=False)
    return df
exl_to_csv(xls)
```

```python
dc=pd.read_csv('C:\\Users\\vivek\\qubryx\\QUB\\s3\\customer.csv')
do=pd.read_csv('C:\\Users\\vivek\\qubryx\\QUB\\s3\\order.csv')
dp=pd.read_csv('C:\\Users\\vivek\\qubryx\\QUB\\s3\\product.csv')

def cleaning_missing_values(dc,do,dp):
    dc.dropna(inplace=True)
    do.dropna(inplace=True)
    dp.dropna(inplace=True)
    return dc, do, dp
dc, do, dp = cleaning_missing_values(dc, do, dp)
print(dc.info())
print(do.info())


dp['price'] = pd.to_numeric(dp['price'], errors='coerce')
print(dp.info())
```

```python
def summary_statistics(dc, do, dp):
    stats = {}
    stats['customer'] = dc.describe(include='all')
    stats['order'] = do.describe(include='all')
    stats['product'] = dp.describe(include='all')

    combined_stats = pd.concat(stats, axis=1)
    combined_stats.to_csv('summary_statistics.csv')

    print(combined_stats)
    return combined_stats
summary_statistics(dc, do, dp)
```

```python
def group_by_category(dp):
    category_summary = dp.groupby('product_category').agg({
        'price': ['sum'],
    }).round(2)

    category_summary.columns = ['price_total']
    category_summary.to_csv('category_summary.csv')

    print(category_summary)
    return category_summary

group_by_category(dp)
```

```python
def merging_data(dc, do, dp):
    do = do.rename(columns={'price': 'order_price'})
    dp = dp.rename(columns={'price': 'product_price'})
    
    merged_data = do.merge(dc, on='customer_id', how='inner')
    merged_data = merged_data.merge(dp, on='product_id', how='inner')
    
    merged_data.to_csv('merged_data.csv', index=False)
    print(merged_data.head())
    return merged_data

merging_data(dc, do, dp)
```

```python
def csv_to_excel_multiple_sheets():

    with pd.ExcelWriter('cleaned_data.xlsx', engine='openpyxl') as writer:
        dc.to_excel(writer, sheet_name='customer', index=False)
        do.to_excel(writer, sheet_name='order', index=False)
        dp.to_excel(writer, sheet_name='product', index=False)
csv_to_excel_multiple_sheets()
```

```python
def upload_to_s3(file_name, bucket, object_name):
    session = boto3.Session(region_name='us-east-1')
    s3_client = session.client("s3")
    
    file_path = f"C:\\Users\\vivek\\qubryx\\QUB\\s3\\{file_name}"
    
    s3_client.upload_file(file_path, bucket, object_name)
    print(f"Successfully uploaded {file_name} to s3://{bucket}/{object_name}")

upload_to_s3('cleaned_data.xlsx', 'qub0013', 'demo/cleaned_data.xlsx')
```
