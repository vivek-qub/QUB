```python
from pypdf import PdfReader
import pandas as pd

reader = PdfReader("/Users/QubryxVivek/QUB/DE/test1.pdf")

pages = reader.pages[0] 

text = pages.extract_text()

print(text)
```

```python
import pandas as pd
from IPython.display import HTML, display

# Parse key-value pairs from text (split by ':')
d = {}
for line in text.splitlines():
    line = line.strip()
    if ':' in line:
        key, val = line.split(':', 1)  # split only on first ':'
        d[key.strip()] = val.strip()

# Create single-row DataFrame from the dictionary
if d:
    df = pd.DataFrame([d])
else:
    df = pd.DataFrame()

# Display as HTML table without index
display(HTML(df.to_html(index=False)))
```
