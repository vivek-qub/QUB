```python
import pdfplumber

def extract_text(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        all_text = []
        for page in pdf.pages:
            text = page.extract_text() + '\n'
            if text:
                all_text.append(text)
    return '\n'.join(all_text)

pdf_path = ("/Users/QubryxVivek/QUB/DE/DE/test1.pdf")

text = extract_text(pdf_path)

print(text)
```

```python
import pandas as pd
import re
from IPython.display import HTML, display

def text_to_df(text):
    d = {}
    
    patterns = [
        r'([A-Za-z0-9\s]+?)\s*:\s*(.+?)(?=\n[A-Z]|\n|\Z)',
        r'([A-Za-z0-9\s]+?)\s*=\s*(.+?)(?=\n[A-Z]|\n|\Z)',
        r'([A-Za-z\s]+?)\s{3,}(.+?)(?=\n[A-Z]|\n|\Z)',
        r'([A-Z\s]+?)\s*:\s*(.+?)(?=\n[A-Z]|\n|\Z)',
        r'([A-Za-z0-9\s]+?)\s*-\s*(.+?)(?=\n[A-Z]|\n|\Z)',
        r'([A-Za-z0-9\s]+?)\s*/\s*(.+?)(?=\n[A-Z]|\n|\Z)',
        r'^([A-Z][A-Za-z\s]{2,}?)\s+(.+?)$',
        r'(Date|DOB|Due\s*Date)\s*:?\s*([0-9/\-\s]+)',
        r'(Name|Customer\s*Name|Client)\s*:?\s*([A-Za-z\s]+)',
        r'(REMARKS)\s*:\s*(.+)',  
        r'^([A-Z\s/]+?)\s+(.+)$',  
    ]
        
    for line_num, line in enumerate(text.splitlines()):
        line = line.strip()
        if not line or len(line) < 3:
            continue
            
        matched = False
        for pattern in patterns:
            match = re.match(pattern, line, re.MULTILINE)
            if match:
                key, val = match.groups()
                key = re.sub(r'[^\w\s]', '_', key.strip())
                val = val.strip()
                
                if val and len(val) > 1:
                    d[key] = val
                    print(f" Line {line_num}: '{key}' → '{val[:50]}...'")
                    matched = True
                break
    
    if d:
        df = pd.DataFrame([d])
        print(f"\n Extracted {len(d)} key-value pairs!")
        display(HTML(df.to_html(index=False, escape=False, classes='table table-striped')))
        return df


df = text_to_df(text)

```
